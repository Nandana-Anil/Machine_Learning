```python
import matplotlib.pyplot as plt
x=[1,2,3,4]
y=[8,9,10,11]
plt.plot(x,y)
plt.show
```

```python
import numpy as np
x=np.array([1,2,6,8])
y=np.array([3,8,1,10])

plt.plot(x,y)
plt.show()
```

```python
x1=np.array([3,8,1,10,5,7])
plt.plot(x1)
```

```python
x=[5,2,3,4]
y=[8,9,10,11]

plt.plot(x,y,'s')  #marker  * + o s
```

```python
x1=np.array([1,2,3,4,5,6])
y1=np.array([3,8,1,1,5,6])
plt.plot(x1,y1,marker='o')
```

```python
plt.plot(x1,y1,'*:r') #marker *, line dotted, color r

#lines: solid(-), dotted(:) dashed()
#colors : k,y,c
```

```python
#marker properties

plt.plot(x1,y1,marker='o',ms=12,mfc='r',mec='g',linestyle='dashed',linewidth=4,color='k') 
#ms ===> marker size
#mfc ===> marker face color
#mec ===> marker edge color
```
