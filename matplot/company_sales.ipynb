```python
import numpy as np 
import pandas as pd 

df=pd.read_csv('company_sales_data.csv')
df
```

```python
df.columns
```

```python
#1. Read Total profit of all months and show it using a line plot
import matplotlib.pyplot as plt

profit=df['total_profit']
month=df['month_number']

plt.plot(month,profit,'o--k')
plt.xlabel('Month Number', fontsize=10)
plt.ylabel('Total Profit', fontsize=10)
plt.title('Total profit by month')
```

```python
# 2. Get total profit of all months and show line plot with the following Style properties
#  Generated line plot must include following Style properties:
#  Line Style dotted and Line-color should be red
#  Show legend at the lower right location.
#  X label name = Month Number
#  Y label name = Sold units number
#  Add a circle marker.
#  Line marker color as read
#  Line width should be 3

plt.plot(month,profit,linestyle='dotted',color='r',marker='o',mfc='red',linewidth=3,label='Total Profit')
plt.xlabel('Month Number')
plt.ylabel('Sold Units Number')
plt.xticks(month)
plt.yticks([100000,200000,300000,400000,500000])
plt.legend(loc='lower right')
```

```python
# 3. Read all product sales data and show it  using a multiline plot

facecream=df['facecream']
facewash=df['facewash']
tooth=df['toothpaste']
bath=df['bathingsoap']
sham=df['shampoo']
mois=df['moisturizer']

plt.plot(month,facecream,marker='s',color='g',label='Face cream sales')
plt.plot(month,facewash,marker='s',color='r',label='Facewash sales')
plt.plot(month,tooth,marker='s',color='k',label='Toothpaste sales')
plt.plot(month,bath,marker='s',color='y',label='Bathingsoap sales')
plt.plot(month,sham,marker='s',color='m',label='Shampoo sales')
plt.plot(month,mois,marker='s',color='c',label='Moisturizer sales')
plt.xlabel('Month number')
plt.ylabel('Total sales of each category')
plt.legend(loc='upper right')
```

```python
# 4.Read toothpaste sales data of each month and show it using a scatterplot

plt.scatter(month,tooth)
```

```python
# 5.Read face cream and facewash product sales data and show it using the bar chart with grid line

plt.bar([a-0.25 for a in month], facecream, width=0.25,align='edge',label='Face cream sales')
plt.bar([a+0.25 for a in month], facewash, width=0.25,align='edge',label='Face wash sales')
plt.xlabel('Month number')
plt.ylabel('Profit')
plt.title('Facecream & Facewash')
plt.legend(loc='upper right')
plt.tight_layout()
plt.xticks(month)
plt.grid()
```

```python
# 6.Read the total profit of each month and show it using the histogram to see the most common profit ranges

plt.hist(profit, color='skyblue', edgecolor='black')
plt.title('Distribution of Total Profit by Month', fontsize=14)
plt.xlabel('Profit Range', fontsize=12)
plt.ylabel('Frequency', fontsize=12)
plt.grid()
plt.show()
```

```python
# 7.Calculate total sale data for last year for each product and show it using a Pie chart

labels=['Facecream','Facewash','Bathing soap','Toothpaste','Shampoo','Moisturiser']
sales=[df['facecream'].sum(),df['facewash'].sum(),df['bathingsoap'].sum(),df['toothpaste'].sum(),df['shampoo'].sum(),df['moisturizer'].sum()]
plt.pie(sales,labels=labels,autopct='%1.01f%%')
plt.legend(loc='upper right')
plt.tight_layout()
```

```python
# 8. Read Bathing soap facewash of all months and display it using the Subplot

plt.subplot(2, 1, 1)
plt.plot(df['month_number'], bath, marker='o', linestyle='-', color='blue')
plt.title('Sales Data of Bathing Soap', fontsize=14)
plt.xlabel('Month Number', fontsize=12)
plt.ylabel('Sales Units', fontsize=12)
plt.grid()

plt.subplot(2, 1, 2)
plt.plot(df['month_number'], facewash, marker='o', linestyle='-', color='green')
plt.title('Sales Data of Facewash', fontsize=14)
plt.xlabel('Month Number', fontsize=12)
plt.ylabel('Sales Units', fontsize=12)
plt.grid()

plt.tight_layout()
plt.show()
```
