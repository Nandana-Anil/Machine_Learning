```python
import numpy as np 
import pandas as pd 

df=pd.read_csv('customer1.csv')
df
```

```python
df.columns
```

```python
df1=df.groupby('loc') ['loc'].count().sort_values(ascending=False)
df1
```

```python
import matplotlib.pyplot as plt

plt.bar(df1.index,df1.values)
plt.xlabel('Location')
plt.ylabel('Count')
plt.title('Customer count based on Location')
```

```python
plt.pie(df1.values,labels=df1.index,autopct='%1.0f%%')
plt.legend(loc='upper left')
plt.tight_layout()
```
