```python
#Encoding
#cant use One-Hot since no.og columns would end up very large
#get_dummies Encoding

df1=pd.get_dummies(df[['Location','Fuel_Type','Transmission','Owner_Type','Brand']],dtype=int,drop_first=True)
#drop_first ==> drop first column
#dtype=int ==> otherwise false,true comes

df1
```

```python
df1.shape
```

```python
dfe=pd.concat([df,df1],axis=1)
dfe
```

```python
dfe.shape
```

```python
#drop unwanted columns

dfe.drop(['Unnamed: 0','Name','Location','Fuel_Type','Transmission','Owner_Type','New_Price','Brand'],inplace=True,axis=1)
#columns that are encoded are also dropped - since they are present as new columns

dfe
```

```python
#remove units
#change null values in columns to 0(as object)

dfe['Mileage'] = dfe['Mileage'].str.replace('km/kg','')
dfe['Mileage'] = dfe['Mileage'].str.replace('kmpl','')
dfe['Engine'] = dfe['Engine'].str.replace('CC','')
dfe['Power'] = dfe['Power'].str.replace('bhp','')
dfe['Mileage']=dfe['Mileage'].str.replace('null','0')
dfe['Power']=dfe['Power'].str.replace('null','0')
dfe['Engine']=dfe['Engine'].str.replace('null','0')
dfe
```

```python
#converting datatype - obj -> float

dfe['Mileage'] = dfe['Mileage'].astype(float)
dfe['Engine'] = dfe['Engine'].astype(float)
dfe['Power'] = dfe['Power'].astype(float)
dfe
```

```python
dfe.dtypes
```

```python
#Consider 0s in Mileage, Power & Engine as missing values and replace

dfe.loc[dfe.Mileage==0,'Mileage']=np.NaN
dfe.loc[dfe.Power==0,'Power']=np.NaN
dfe.loc[dfe.Engine==0,'Engine']=np.NaN
dfe
```

```python
dfe['Engine']=dfe['Engine'].fillna(dfe['Engine'].mean())
dfe['Power']=dfe['Power'].fillna(dfe['Power'].mean())
dfe['Mileage']=dfe['Mileage'].fillna(dfe['Mileage'].mean())
dfe['Seats']=dfe['Seats'].fillna(dfe['Seats'].mode()[0])
```
