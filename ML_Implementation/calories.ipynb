```python
numeric_cols=df.select_dtypes(include=[np.number]).columns #boxplot can only be done on numeric cols

plt.figure(figsize=(12,8))
sb.boxplot(data=df[numeric_cols])
plt.title("Boxplot of all features")
plt.show()
```

```python
#outlier removal

outlier_cols=['Height',"Weight","Heart_Rate","Body_Temp","Calories"]

def rem_outliers(df,columns):
    for col in columns:

        Q1=df[col].quantile(0.25)
        Q3=df[col].quantile(0.75)
        IQR=Q3-Q1

        low=Q1-1.5*IQR
        upp=Q3+1.5*IQR

        df=df[(df[col] >= low) & (df[col]<=upp)]

    return df
df_cleaned=rem_outliers(df,outlier_cols)
```

```python
corr_matrix=df_cleaned.corr() #correlation
plt.figure(figsize=(8,6))
sb.heatmap(corr_matrix,annot=True,cmap="coolwarm",fmt=".2f",linewidths=0.5) 
#annot = True => to view values within cells

plt.title("Heatmap")
plt.show()
```

```python
#low correlation b/w height, gender and calories => drop 

df_selected=df_cleaned.drop(columns=["Height","Gender"])

corr_matrix2=df_selected.corr() 
plt.figure(figsize=(8,6))
sb.heatmap(corr_matrix2,annot=True,cmap="coolwarm",fmt=".2f",linewidths=0.5) 
plt.title("Heatmap")
plt.show()
```

```python
x=df_selected.iloc[:,:-1]
x
```

```python
y=df_selected.iloc[:,-1]
y
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.30,random_state=1)
x_train
```

```python
#model creation

from sklearn.linear_model import LinearRegression
model=LinearRegression()
model.fit(x_train,y_train)

y_pred=model.predict(x_test)
y_pred
```

```python
df2=pd.DataFrame({'Actual_value':y_test,'Predicted_value':y_pred,'Difference':y_test-y_pred})
df2
```

```python
from sklearn.metrics import mean_absolute_percentage_error,r2_score

mape=mean_absolute_percentage_error(y_test,y_pred)
mape
```

```python
sc=r2_score(y_test,y_pred)
sc
```
