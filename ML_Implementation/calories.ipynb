```python
df['Age']=df['Age'].fillna(df['Age'].median())
df['Weight']=df['Weight'].fillna(df['Age'].mean())
df.isna().sum()
```

```python
df.drop(['User_ID'],axis=1,inplace=True)
df
```

```python
from sklearn.preprocessing import LabelEncoder
lab=LabelEncoder()
df['Gender']=lab.fit_transform(df['Gender'])
df
```

```python
gender_count=df['Gender'].value_counts()
gender_labels=['Male','Female']

plt.figure(figsize=(8,5)) #8 inch wide, 5 inch ht
plt.pie(gender_count,labels=gender_labels,autopct='%1.1f%%',colors=['pink','coral'])
plt.title("Gender Distribution")
plt.show()
```

```python
sb.scatterplot(x='Height',y='Weight',data=df)
plt.show() #shows linear relation
```

```python
plt.figure(figsize=(8,6))
sb.barplot(x='Duration',y='Calories',data=df,estimator=np.mean,palette="viridis",hue='Duration')
plt.title("Average Calories Burnt vs Duration (in minutes)")
plt.xlabel("Duration")
plt.ylabel("Average Calories burnt")
plt.xticks(rotation=45) #to improve readability, xlabels rotated
plt.show()
```

```python
plt.figure(figsize=(8,6))
sb.scatterplot(x="Duration",y="Body_Temp",data=df,hue="Duration")
plt.xlabel("Duration in mins")
plt.ylabel("Body Temperature")
plt.show()
```
